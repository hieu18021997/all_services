```python
import pandas as pd
import numpy as np
from vnstock import *
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from datetime import datetime, date, timedelta
START_DATE = '2013-01-01'
END_DATE = '2023-10-12'
SYSBOL = 'HPG'
```

```python
# Lấy dữ liệu giá cả cổ phiếu
# data = pd.read_csv('data.csv', index_col='Date')
df = stock_historical_data(SYSBOL, START_DATE, END_DATE, "1D", "stock")
df.set_index('time')
df.head()

# Tính toán EMA 9 và EMA 50
ema_9 = df['close'].ewm(span=9, min_periods=8).mean()
ema_50 = df['close'].ewm(span=50, min_periods=49).mean()

# Xác định các tín hiệu giao dịch
signals = []
for i in range(len(df)):
    if ema_9[i] > ema_50[i] and ema_9[i - 1] < ema_50[i - 1]:
        signals.append('Buy')
    elif ema_9[i] < ema_50[i] and ema_9[i - 1] > ema_50[i - 1]:
        signals.append('Sell')
    else:
        signals.append('None')

# In dữ liệu
print(df)
print(signals)
```

```python
# Tạo chart
fig = go.Figure()
fig.add_trace(go.Scatter(x=df['time'], y=df['close'], name='Giá'))
fig.add_trace(go.Scatter(x=df['time'], y=ema_9, line_color='green', name='EMA 9'))
fig.add_trace(go.Scatter(x=df['time'], y=ema_50, line_color='red', name='EMA 50'))

# Xác định các tín hiệu giao dịch
for i in range(len(df)):
    if ema_9[i] > ema_50[i] and ema_9[i - 1] < ema_50[i - 1]:
        fig.add_trace(go.Scatter(x=[df['time'][i]], y=[df['close'][i]], marker_color='green', name='Mua'))
    elif ema_9[i] < ema_50[i] and ema_9[i - 1] > ema_50[i - 1]:
        fig.add_trace(go.Scatter(x=[df['time'][i]], y=[df['close'][i]], marker_color='red', name='Bán'))

# Thêm tiêu đề và chú thích
fig.update_layout(title='Chiến lược cross EMA 9 và EMA 50 cho cổ phiếu HPG', xaxis_title='Ngày', yaxis_title='Giá')

# Hiển thị chart
fig.show()
```
