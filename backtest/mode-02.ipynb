```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from vnstock import *
from sklearn.cluster import DBSCAN
import plotly.graph_objects as go
```

```python
START_DATE = '2013-01-01'
END_DATE = '2023-10-12'
SYSBOL = 'SSI'
```

```python
# Lấy dữ liệu giá cả cổ phiếu
# data = pd.read_csv('data.csv', index_col='Date')
df = stock_historical_data(SYSBOL, START_DATE, END_DATE, "1D", "stock")
df.set_index('time')
df.head()
```

```python
data = df
data = data.rename(columns={"time" : "Date", "close" : "Price"})
close = data['Price']
close = close.fillna(0)
# data = data.to_numpy()
```

```python
# Huấn luyện mô hình
kmeans = KMeans(n_clusters=3)
array = close.to_numpy().reshape(-1,1)
kmeans.fit(array)

# Thử nghiệm mô hình
labels = kmeans.predict(array)
```

```python
print(len(labels))
print(labels)

for label in range(3):
  points = data[labels == label]
print(points)
```

```python
# Tạo figure
fig = go.Figure()

# Thêm dữ liệu giá cổ phiếu
fig.add_trace(go.Scatter(x=data['Date'], y=data['Price'], mode="lines"))

# Thêm dữ liệu phân cụm
for label in range(3):
  points = data[labels == label]
  fig.add_trace(go.Scatter(x=points['Date'], y=points['Price'], mode="markers", marker_color=f"rgb({label * 255 // 3}, {(label + 1) * 255 // 3}, {(label + 2) * 255 // 3})"))

# Tùy chỉnh figure
fig.update_layout(title="K-means clustering", xaxis_title="Date", yaxis_title="Price")

# Hiển thị figure
fig.show()
```
