```python
stock_historical_data(SYSBOL, START_DATE, END_DATE, "1D")
```

```python
df = stock_historical_data(SYSBOL, START_DATE, END_DATE, "1D", "stock")
fig = candlestick_chart(df, ma_periods=[50,200], show_volume=False, reference_period=300, figure_size=(15, 8), 
                        title=f'{SYSBOL} - Candlestick Chart with MA and Volume', x_label='Date', y_label='Price', 
                        colors=('lightgray', 'gray'), reference_colors=('black', 'blue'))
fig
```

```python
df = bollinger_bands(df, window=20, num_std_dev=2)
fig = bollinger_bands_chart(df, use_candlestick=True, show_volume=True, 
                            fig_size=(15, 8), chart_title='Bollinger Bands Chart', xaxis_title='Date', yaxis_title='Price', 
                            bollinger_band_colors=('gray', 'orange', 'gray'), volume_colors=('#00F4B0', '#FF3747'))
fig.show()
```
